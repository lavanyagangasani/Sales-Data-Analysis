```python
# ============================================
# SALES DATA ANALYSIS PROJECT
# Tech Stack: Python, Pandas, Matplotlib
# ============================================

# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# --------------------------------------------
# Step 1: Read the CSV file
# --------------------------------------------
df = pd.read_csv("sales_data.csv")

# --------------------------------------------
# Step 2: Display first 5 rows
# --------------------------------------------
print("First 5 Rows:")
print(df.head())

# --------------------------------------------
# Step 3: Display dataset information
# --------------------------------------------
print("\nDataset Information:")
print(df.info())

# --------------------------------------------
# Step 4: Check missing values
# --------------------------------------------
print("\nMissing Values:")
print(df.isnull().sum())

# --------------------------------------------
# Step 5: Remove missing values
# --------------------------------------------
df = df.dropna()

# --------------------------------------------
# Step 6: Remove duplicate rows
# --------------------------------------------
df = df.drop_duplicates()

# --------------------------------------------
# Step 7: Convert Date column into Date format
# --------------------------------------------
df["Date"] = pd.to_datetime(df["Date"])

# --------------------------------------------
# Step 8: Create Month column
# --------------------------------------------
df["Month"] = df["Date"].dt.month_name()

# --------------------------------------------
# Step 9: Basic Statistics
# --------------------------------------------
print("\n========== SALES SUMMARY ==========")

print("Total Sales:", df["Sales"].sum())

print("Average Sales:", df["Sales"].mean())

print("Highest Sales:", df["Sales"].max())

print("Lowest Sales:", df["Sales"].min())

# --------------------------------------------
# Step 10: Monthly Sales
# --------------------------------------------
monthly_sales = df.groupby("Month")["Sales"].sum()

print("\nMonthly Sales:")
print(monthly_sales)

# --------------------------------------------
# Step 11: Monthly Sales Bar Chart
# --------------------------------------------
plt.figure(figsize=(10,5))

monthly_sales.plot(kind="bar")

plt.title("Monthly Sales")

plt.xlabel("Month")

plt.ylabel("Sales")

plt.xticks(rotation=45)

plt.show()

# --------------------------------------------
# Step 12: Sales Trend Line Chart
# --------------------------------------------
plt.figure(figsize=(12,5))

plt.plot(df["Date"], df["Sales"], marker="o")

plt.title("Sales Trend Over Time")

plt.xlabel("Date")

plt.ylabel("Sales")

plt.grid(True)

plt.show()

# --------------------------------------------
# Step 13: Sales Distribution Histogram
# --------------------------------------------
plt.figure(figsize=(8,5))

plt.hist(df["Sales"], bins=20)

plt.title("Sales Distribution")

plt.xlabel("Sales")

plt.ylabel("Frequency")

plt.show()

# --------------------------------------------
# Step 14: Save Cleaned Dataset
# --------------------------------------------
df.to_csv("cleaned_sales_data.csv", index=False)

print("\nCleaned dataset saved as cleaned_sales_data.csv")

# --------------------------------------------
# Step 15: Generate Report
# --------------------------------------------
report = f"""
SALES DATA ANALYSIS REPORT
==========================

Total Sales : {df['Sales'].sum()}

Average Sales : {df['Sales'].mean():.2f}

Highest Sales : {df['Sales'].max()}

Lowest Sales : {df['Sales'].min()}

Total Records : {len(df)}
"""

with open("Sales_Report.txt", "w") as file:
    file.write(report)

print("\nSales report generated successfully.")

# --------------------------------------------
# Step 16: Display Final Dataset
# --------------------------------------------
print("\nFinal Dataset:")
print(df.head())
```
